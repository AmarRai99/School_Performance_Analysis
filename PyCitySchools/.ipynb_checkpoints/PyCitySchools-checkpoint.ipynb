```python
#Input dependencies and setup(import).

import os
import csv as c
import pandas as pd
import numpy as np
```

```python
#Load the files for school data and students data

schools_data_path = "/Users/amarrai/Desktop/pandas-challenge/Resources/schools_complete.csv"
students_data_path = "/Users/amarrai/Desktop/pandas-challenge/Resources/students_complete.csv"
```

```python
#Read the school and students data files 
#Store the data in a Pandas Data Frame

schools_data = pd.read_csv(schools_data_path)
students_data = pd.read_csv(students_data_path)

#Merge the data into one data set
complete_school_data = pd.merge(students_data, schools_data, how="left", on=["school_name", "school_name"])

#Output
complete_school_data
```

# District Summary

* Calculate the Total number of schools
* Calculate the Total number of students
* Calculate the Total budget
* Calculate the Average math score
* Calculate the Average reading score
* Calculate the percentage of students that have passed math (70 or more)
* Calculate the precentage of students that have passed reading (the percentage of students who passed reading)
* Calculate the percentage of students who passed math AND reading
* Create a DataFrame for the results

```python
# Calculating the total number of shcools

total_number_schools = len(complete_school_data["School ID"].unique())

# Output
total_number_schools
```

```python
# Calculating the total number of students

total_number_students = len(complete_school_data["Student ID"].unique())

# Output 
total_number_students
```

```python
# Calculating the Total Budget

total_budget = schools_data ["budget"].sum()

# Output 

total_budget
```

```python
# Calculating average math score

average_math_score = students_data["math_score"].mean()

# Output
average_math_score
```

```python
# Calculate average reading score

average_reading_score = students_data["reading_score"].mean()

# Output
average_reading_score
```

```python
# Percentage of students who passed math (70 or more)

pass_math_score = (complete_school_data[complete_school_data["math_score"]>=70]["student_name"].count()/total_number_students)*100

# Output
pass_math_score
```

```python
# Percentage of students who passed reading

pass_reading_score = (complete_school_data [complete_school_data["reading_score"] >=70]["student_name"].count()/total_number_students)*100

# Output
pass_reading_score
```

```python
# Percentage of students who passed math and reading

# Passing both reading and math.
passing_math_reading = complete_school_data[(complete_school_data["math_score"] >= 70)
                                               & (complete_school_data["reading_score"] >= 70)]

# Number of students that passed both reading and math.
overall_passed = passing_math_reading["student_name"].count()


#Calculating the overall passing percentage.
Overall_score = (overall_passed/total_number_students) * 100

#Output
Overall_score
```

```python
#Creating a dataframe to hold the results.
district_results = [{"Total Schools": total_number_schools, 
            "Total Students": total_number_students, 
            "Total Budget": total_budget, 
            "Average Math Score":  round(average_math_score,2), 
            "Average Reading Score":  round(average_reading_score,2), 
           "% Passing Math": round(pass_math_score,2),
           "% Passing Reading": round(pass_reading_score,2),
            "% Overall Passing Rate": round(Overall_score,2)}]

district_summary_table = pd.DataFrame(district_results)

#Format total students to have comma
district_summary_table["Total Students"] = district_summary_table["Total Students"].map("{:,}".format)

#Format total budjet to have currency sign and commas.
district_summary_table["Total Budget"] = district_summary_table["Total Budget"].map("${:,.2f}".format)

#Format average/percentages to have 2.d.p
district_summary_table["Average Math Score"] = district_summary_table["Average Math Score"].map("{:,.2f}".format)
district_summary_table["Average Reading Score"] = district_summary_table["Average Reading Score"].map("{:,.2f}".format)
district_summary_table["% Passing Math"] = district_summary_table["% Passing Math"].map("{:,.2f}".format)
district_summary_table["% Passing Reading"] = district_summary_table["% Passing Reading"].map("{:,.2f}".format)
district_summary_table["% Overall Passing Rate"] = district_summary_table["% Overall Passing Rate"].map("{:,.2f}".format)
```

```python
#Displaying the table
district_summary_table
```

# School Summary

* Create an overview table that summarizes key metrics about each school, including:

* School Name
* School Type
* Total Students
* Total School Budget
* Per Student Budget
* Average Math Score
* Average Reading Score
* % Passing Math
* % Passing Reading
* % Overall Passing (The percentage of students that passed math and reading.)
* Create a dataframe to hold the above results

```python
#Group and determine the school name and type
school_groups = complete_school_data.groupby(['school_name'])
```

```python
#Calculating school types
type_of_school = school_groups['type'].first()

#Calculating total Students
total_number_students = school_groups.size()
total_number_students = complete_school_data["school_name"].value_counts()

#Total budget per school
total_budget = school_groups['budget'].first()

#Total school budget
budget_per_school = complete_school_data.groupby(["school_name"]).mean()["budget"]

#Per student budget
capita_per_school = budget_per_school / total_number_students

budget_per_student = total_budget / total_number_students

# Average Maths score
average_math_score = school_groups['math_score'].mean()

# Average reading score
average_reading_score = school_groups['reading_score'].mean()

# % passing math
groups_passing_maths = complete_school_data[complete_school_data['math_score']>=70].groupby(['school_name']).size()

percent_passing_maths = (groups_passing_maths / total_number_students) *100

# % passing reading
groups_passing_reading = complete_school_data[complete_school_data['reading_score']>=70].groupby(['school_name']).size()
percent_passing_reading = (groups_passing_reading/total_number_students)*100

# % Overall passing

percent_passing_math_reading = complete_school_data[(complete_school_data["reading_score"] >= 70) & (complete_school_data["math_score"] >= 70)]

percent_passing_math_reading = percent_passing_math_reading.groupby(["school_name"]).count()["student_name"]

                                           
percent_overall_passing = percent_passing_math_reading / total_number_students * 100


```

```python
# Creating DataFrame for Results

schools_results = { "Type Of School": type_of_school,
                  "Total Number Of Students": total_number_students,
                  "Total School Budget": total_budget,
                  "Budget Per Student": budget_per_student,
                  "Average Math Score": average_math_score,
                  "Average Reading Score": average_reading_score,
                  "% Passing Maths": percent_passing_maths,
                  "% Passing Reading": percent_passing_reading,
                  "% Overall Passing": percent_overall_passing}

results_summary_table = pd.DataFrame(schools_results)


results_summary_table["Total Number Of Students"] = results_summary_table["Total Number Of Students"].map("{:,.0f}".format)
results_summary_table["Total School Budget"] = results_summary_table["Total School Budget"].map("${:,.2f}".format)
results_summary_table["Budget Per Student"] = results_summary_table["Budget Per Student"].map("${:,.2f}".format)


results_summary_table.index.name = None
```

# Scores by School Spending

Create a table that breaks down school performance based on average spending ranges (per student). Use your judgment to create four bins with reasonable cutoff values to group school spending. Include the following metrics in the table:

* Average math score
* Average reading score
* % passing math (the percentage of students who passed math)
* % passing reading (the percentage of students who passed reading)
* % overall passing (the percentage of students who passed math AND reading)

```python
# Group names and spending bin
bins = [0, 580, 640, 675, 690]
group_names = ["<$580", "$581-640", "$641-675", "$676-690"]

results_summary_table["Spending Ranges (per student)"] = pd.cut(budget_per_student, bins, labels = group_names)

results_summary_table
 
```

```python
# Calculating the averages for the bins

maths_bin = results_summary_table.groupby(["Spending Ranges (per student)"]).mean()["Average Math Score"]
reading_bin = results_summary_table.groupby(["Spending Ranges (per student)"]).mean()["Average Reading Score"]
maths_pass_bin = results_summary_table.groupby(["Spending Ranges (per student)"]).mean()["% Passing Maths"]
reading_pass_bin  = results_summary_table.groupby(["Spending Ranges (per student)"]).mean()["% Passing Reading"]
overall_pass_bin = results_summary_table.groupby(["Spending Ranges (per student)"]).mean()["% Overall Passing"]

# Create the DataFrame

spending_bins_summary = pd.DataFrame({
   "Average Maths Score": maths_bin,
   "Average Reading Score": reading_bin,
   "% Passing Maths": maths_pass_bin,
    "% Passing Reading": reading_pass_bin,
    "% Overall Passing": overall_pass_bin   
})

# Format to 2 decimal places

spending_bins_summary["Average Maths Score"] = spending_bins_summary["Average Maths Score"].map("{:.2f}%".format)
spending_bins_summary["Average Reading Score"] = spending_bins_summary["Average Reading Score"].map("{:.2f}%".format)
spending_bins_summary["% Passing Maths"] = spending_bins_summary["% Passing Maths"].map("{:.2f}%".format)
spending_bins_summary["% Passing Reading"] = spending_bins_summary["% Passing Reading"].map("{:.2f}%".format)
spending_bins_summary["% Overall Passing"] = spending_bins_summary["% Overall Passing"].map("{:.2f}%".format)


spending_bins_summary




```

# Scores by School Size

Create a table that breaks down school performance based on school size (small, medium, or large).

```python
# Spending bins and group names

size_of_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]


# Create copy of table being used

size_of_schools = results_summary_table.copy()

# Schools sizes 
size_of_schools["School Size"] = pd.cut(total_number_students, bins = size_of_bins, labels = size_labels)

# Calculate Averages

size_maths_scores = size_of_schools.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = size_of_schools.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = size_of_schools.groupby(["School Size"]).mean()["% Passing Maths"]
size_passing_reading = size_of_schools.groupby(["School Size"]).mean()["% Passing Reading"]
overall_passing_size = size_of_schools.groupby(["School Size"]).mean()["% Overall Passing"]

# Create DataFrame

size_of_schools_summary = pd.DataFrame({
    "Average Math Score": size_maths_scores,
   "Average Reading Score": size_reading_scores,
   "% Passing Maths": size_passing_math,
    "% Passing Reading": size_passing_reading,
    "% Overall Passing": overall_passing_size   
    })

size_of_schools_summary





```
